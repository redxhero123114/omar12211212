# 🌃 LoLI-Street — Low-Light Image Enhancement Pipeline

> **Dataset:** LoLI-Street (tanvirnwu — Kaggle)  
> **Model:** U-Net Encoder–Decoder  
> **Framework:** TensorFlow / Keras  
> **Runtime:** GPU (T4 recommended)

---
### Pipeline Overview
```
Raw Low-Light Image
       │
       ▼
  Preprocessing  →  Resize │ Denoise │ CLAHE │ Canny Edges
       │
       ▼
  U-Net Model    →  Encoder (32→64→128→256→512) + Decoder with Skip Connections
       │
       ▼
  Enhanced Image (256×256×3)
```

## ⚙️ Cell 3 — Configuration

````python
import os

# ── Dataset path (update if your path differs) ───────────────────────────────
DATASET_PATH = "/root/.cache/kagglehub/datasets/tanvirnwu/loli-street-low-light-image-enhancement-of-street/versions/1/LoLI-Street Dataset"

# ── Hyperparameters ───────────────────────────────────────────────────────────
IMAGE_SIZE    = (256, 256)   # All images resized to this
MAX_IMAGES    = 7000         # Max training samples
BATCH_SIZE    = 16
EPOCHS        = 50
LEARNING_RATE = 1e-4
MODEL_PATH    = "best_model.keras"

# ── Preprocessing parameters ─────────────────────────────────────────────────
CLAHE_CLIP    = 2.0
CLAHE_GRID    = (8, 8)
CANNY_LOW     = 50
CANNY_HIGH    = 150

print("✅ Config ready")
print(f"   Image size : {IMAGE_SIZE}")
print(f"   Max images : {MAX_IMAGES}")
print(f"   Batch size : {BATCH_SIZE}")
print(f"   Epochs     : {EPOCHS}")
````

## 📚 Cell 4 — Imports

````python
import glob
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, Model
from tensorflow.keras.callbacks import (
    ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
)

print("✅ All imports OK")
print("   TensorFlow version:", tf.__version__)
print("   GPU available:", tf.config.list_physical_devices('GPU'))
````

## 🔧 Cell 5 — Preprocessing Function (CLAHE + Canny)

````python
def enhance_image(img_bgr):
    """
    Full preprocessing pipeline for one BGR image.

    Steps
    -----
    1. BGR -> RGB + resize to IMAGE_SIZE
    2. Gaussian blur (5x5)  — noise removal
    3. RGB -> LAB
    4. CLAHE on L channel   — adaptive contrast enhancement
    5. LAB -> RGB
    6. Canny edge detection on grayscale version
    7. Normalise both to [0, 1]

    Returns
    -------
    enhanced_norm : float32 (H, W, 3)  — enhanced RGB image
    edges_norm    : float32 (H, W)     — binary edge map
    """
    # Step 1 — resize
    img_rgb     = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    img_resized = cv2.resize(img_rgb, IMAGE_SIZE)

    # Step 2 — denoise
    denoised = cv2.GaussianBlur(img_resized, (5, 5), 0)

    # Step 3-5 — CLAHE in LAB space
    lab        = cv2.cvtColor(denoised, cv2.COLOR_RGB2LAB)
    l, a, b    = cv2.split(lab)
    clahe      = cv2.createCLAHE(clipLimit=CLAHE_CLIP, tileGridSize=CLAHE_GRID)
    cl         = clahe.apply(l)
    merged     = cv2.merge((cl, a, b))
    enhanced   = cv2.cvtColor(merged, cv2.COLOR_LAB2RGB)

    # Step 6 — Canny edges
    gray   = cv2.cvtColor(enhanced, cv2.COLOR_RGB2GRAY)
    gray   = cv2.GaussianBlur(gray, (3, 3), 0)
    edges  = cv2.Canny(gray, CANNY_LOW, CANNY_HIGH)

    # Step 7 — normalise
    enhanced_norm = enhanced.astype(np.float32) / 255.0
    edges_norm    = edges.astype(np.float32)    / 255.0

    return enhanced_norm, edges_norm

print("✅ enhance_image() defined")
````

## 📂 Cell 7 — Load & Build Dataset Arrays

````python
def build_dataset(dataset_root):
    """
    Tries to load paired Low/High images for supervised training.
    Falls back to self-supervised mode (CLAHE output as target)
    if no High folder is found.

    Returns
    -------
    X : float32 (N, H, W, 3) — low-light inputs
    Y : float32 (N, H, W, 3) — enhanced targets
    """
    low_files  = sorted(glob.glob(os.path.join(dataset_root, "Train", "Low",  "*.*")))
    high_files = sorted(glob.glob(os.path.join(dataset_root, "Train", "High", "*.*")))

    X_list, Y_list = [], []

    if low_files and high_files:
        # ── Supervised mode ──────────────────────────────────────────────────
        print(f"📁 Supervised mode: {len(low_files)} pairs found")
        n = min(len(low_files), len(high_files), MAX_IMAGES)
        for i, (lp, hp) in enumerate(zip(low_files[:n], high_files[:n])):
            low  = cv2.imread(lp)
            high = cv2.imread(hp)
            if low is None or high is None:
                continue
            x = cv2.resize(cv2.cvtColor(low,  cv2.COLOR_BGR2RGB), IMAGE_SIZE).astype(np.float32) / 255.0
            y = cv2.resize(cv2.cvtColor(high, cv2.COLOR_BGR2RGB), IMAGE_SIZE).astype(np.float32) / 255.0
            X_list.append(x)
            Y_list.append(y)
            if (i + 1) % 500 == 0:
                print(f"   Loaded {i+1}/{n}...")
    else:
        # ── Self-supervised mode ─────────────────────────────────────────────
        all_files = sorted(glob.glob(os.path.join(dataset_root, "Train", "*", "*")))[:MAX_IMAGES]
        print(f"📁 Self-supervised mode: {len(all_files)} images found")
        for i, fp in enumerate(all_files):
            img = cv2.imread(fp)
            if img is None:
                continue
            raw      = cv2.resize(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), IMAGE_SIZE).astype(np.float32) / 255.0
            enhanced, _ = enhance_image(img)
            X_list.append(raw)
            Y_list.append(enhanced)
            if (i + 1) % 500 == 0:
                print(f"   Processed {i+1}/{len(all_files)}...")

    X = np.array(X_list, dtype=np.float32)
    Y = np.array(Y_list, dtype=np.float32)
    print(f"\n✅ Dataset ready  →  X: {X.shape}   Y: {Y.shape}")
    return X, Y


X, Y = build_dataset(DATASET_PATH)
````

## 🧠 Cell 8 — Build U-Net Model

````python
def conv_block(x, filters, dropout=0.0):
    """Two Conv2D + BatchNorm + ReLU layers with optional dropout."""
    x = layers.Conv2D(filters, 3, padding="same", use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    if dropout:
        x = layers.Dropout(dropout)(x)
    x = layers.Conv2D(filters, 3, padding="same", use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    return x


def build_unet(input_shape=(256, 256, 3)):
    """
    U-Net for image-to-image enhancement.

    Encoder  : 4 blocks  (32 → 64 → 128 → 256 filters)
    Bottleneck: 512 filters
    Decoder  : 4 blocks with skip connections
    Output   : Conv(3, 1x1) + Sigmoid → RGB [0,1]
    """
    inputs = layers.Input(shape=input_shape)

    # ── Encoder ──────────────────────────────────────────────────────────────
    c1 = conv_block(inputs, 32)
    p1 = layers.MaxPooling2D()(c1)

    c2 = conv_block(p1, 64)
    p2 = layers.MaxPooling2D()(c2)

    c3 = conv_block(p2, 128)
    p3 = layers.MaxPooling2D()(c3)

    c4 = conv_block(p3, 256, dropout=0.3)
    p4 = layers.MaxPooling2D()(c4)

    # ── Bottleneck ────────────────────────────────────────────────────────────
    c5 = conv_block(p4, 512, dropout=0.3)

    # ── Decoder (with skip connections) ───────────────────────────────────────
    u6 = layers.UpSampling2D()(c5)
    u6 = layers.Concatenate()([u6, c4])
    c6 = conv_block(u6, 256)

    u7 = layers.UpSampling2D()(c6)
    u7 = layers.Concatenate()([u7, c3])
    c7 = conv_block(u7, 128)

    u8 = layers.UpSampling2D()(c7)
    u8 = layers.Concatenate()([u8, c2])
    c8 = conv_block(u8, 64)

    u9 = layers.UpSampling2D()(c8)
    u9 = layers.Concatenate()([u9, c1])
    c9 = conv_block(u9, 32)

    # ── Output ────────────────────────────────────────────────────────────────
    outputs = layers.Conv2D(3, 1, activation="sigmoid")(c9)

    return Model(inputs, outputs, name="UNet_Enhancement")


model = build_unet(input_shape=(*IMAGE_SIZE, 3))
model.summary()
````

## 📉 Cell 9 — Loss Functions & Compile

````python
def ssim_loss(y_true, y_pred):
    """1 - SSIM  (maximising SSIM = minimising this loss)."""
    return 1.0 - tf.reduce_mean(
        tf.image.ssim(y_true, y_pred, max_val=1.0)
    )

def combined_loss(y_true, y_pred):
    """0.7 x MAE  +  0.3 x (1 - SSIM)."""
    mae  = tf.reduce_mean(tf.abs(y_true - y_pred))
    ssim = ssim_loss(y_true, y_pred)
    return 0.7 * mae + 0.3 * ssim


model.compile(
    optimizer=tf.keras.optimizers.Adam(LEARNING_RATE),
    loss=combined_loss,
    metrics=["mae", tf.keras.metrics.MeanSquaredError(name="mse")]
)

print("✅ Model compiled")
print(f"   Loss     : 0.7 × MAE + 0.3 × (1 - SSIM)")
print(f"   Optimiser: Adam (lr={LEARNING_RATE})")
````

## 🚀 Cell 10 — Train the Model

````python
callbacks = [
    ModelCheckpoint(
        MODEL_PATH,
        monitor="val_loss",
        save_best_only=True,
        verbose=1
    ),
    EarlyStopping(
        monitor="val_loss",
        patience=8,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.5,
        patience=4,
        min_lr=1e-6,
        verbose=1
    ),
]

history = model.fit(
    X, Y,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    validation_split=0.1,
    callbacks=callbacks,
    shuffle=True,
)

print("\n✅ Training complete!")
print(f"   Best model saved → {MODEL_PATH}")
````

## 📊 Cell 11 — Plot Training Curves

````python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Loss
axes[0].plot(history.history["loss"],     label="Train Loss", linewidth=2)
axes[0].plot(history.history["val_loss"], label="Val Loss",   linewidth=2, linestyle="--")
axes[0].set_title("Combined Loss (0.7×MAE + 0.3×SSIM)", fontsize=13)
axes[0].set_xlabel("Epoch")
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# MAE
axes[1].plot(history.history["mae"],     label="Train MAE", linewidth=2)
axes[1].plot(history.history["val_mae"], label="Val MAE",   linewidth=2, linestyle="--")
axes[1].set_title("Mean Absolute Error", fontsize=13)
axes[1].set_xlabel("Epoch")
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.suptitle("LoLI-Street U-Net — Training History", fontsize=15, fontweight="bold")
plt.tight_layout()
plt.savefig("training_history.png", dpi=150)
plt.show()
print("✅ Saved → training_history.png")
````

## 📐 Cell 12 — Evaluate: PSNR & SSIM

````python
# Predict on full dataset
preds = model.predict(X, batch_size=BATCH_SIZE, verbose=1)
preds = np.clip(preds, 0, 1)

psnr_scores, ssim_scores = [], []

for gt, pr in zip(Y, preds):
    gt_t = tf.constant(gt[None], dtype=tf.float32)
    pr_t = tf.constant(pr[None], dtype=tf.float32)
    psnr_scores.append(float(tf.image.psnr(gt_t, pr_t, max_val=1.0)))
    ssim_scores.append(float(tf.image.ssim(gt_t, pr_t, max_val=1.0)))

mean_psnr = np.mean(psnr_scores)
mean_ssim = np.mean(ssim_scores)

print("\n" + "═"*45)
print("  📊  Evaluation Metrics")
print("═"*45)
print(f"  Mean PSNR  :  {mean_psnr:.2f} dB")
print(f"  Mean SSIM  :  {mean_ssim:.4f}")
print("═"*45)
print("  PSNR > 25 dB  →  Good quality")
print("  SSIM > 0.80   →  Good structure")
````
