```python
import glob
import cv2

path = "/root/.cache/kagglehub/datasets/tanvirnwu/loli-street-low-light-image-enhancement-of-street/versions/1/LoLI-Street Dataset"
files = glob.glob(path + "/Train/*/*", recursive=True)

processed_images = []

# Loop through all image files
for file in files[:7000]:

    # Read image
    img = cv2.imread(file)

    # Convert image from BGR to RGB
    img_rgb = cv2.cvtColor(
        img,
        cv2.COLOR_BGR2RGB
    )

    # Resize image to 256x256
    img_resized = cv2.resize(
        img_rgb,
        (256, 256)
    )

    # Apply Gaussian Blur for noise removal
    denoised = cv2.GaussianBlur(
        img_resized,
        (5, 5),
        0
    )

    # Convert RGB image to LAB color space
    lab = cv2.cvtColor(
        denoised,
        cv2.COLOR_RGB2LAB
    )

    # Split LAB channels
    l, a, b = cv2.split(lab)

    # Create CLAHE object for contrast enhancement
    clahe = cv2.createCLAHE(
        clipLimit=2.0,
        tileGridSize=(8, 8)
    )

    # Apply CLAHE on L channel
    cl = clahe.apply(l)

    # Merge channels back
    merged = cv2.merge((cl, a, b))

    # Convert LAB image back to RGB
    final = cv2.cvtColor(
        merged,
        cv2.COLOR_LAB2RGB
    )

    # Normalize pixel values to range [0,1]
    normalized = final / 255.0

    # Store processed image
    processed_images.append(normalized)

# Print total processed images
print("Total processed images:")
print(len(processed_images))
```
